```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Online_Course_Student_Feedback_Analysis_data.csv")
```

```python
print(df.columns)
print(df.info())
```

```python
print(df.head())
```

```python
df = df[['Course', 'Rating', 'Feedback']]
```

```python
df['Rating'].describe()
```

```python
import matplotlib.pyplot as plt

rating_counts = df['Rating'].value_counts().sort_index()

plt.bar(rating_counts.index, rating_counts.values)
plt.xlabel("Rating")
plt.ylabel("Number of Students")
plt.title("Online Course Rating Distribution")
plt.show()
```

```python
def satisfaction_level(rating):
    if rating >= 4:
        return "High"
    elif rating == 3:
        return "Medium"
    else:
        return "Low"

df.loc[:, 'satisfaction'] = df['Rating'].apply(satisfaction_level)

satisfaction_counts = df['satisfaction'].value_counts()
print(satisfaction_counts)
```

```python
satisfaction_counts = df['satisfaction'].value_counts()

plt.bar(satisfaction_counts.index, satisfaction_counts.values)
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of Students")
plt.title("Student Satisfaction Analysis")
plt.show()
```
